```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = []
y = []
archivo = open("linear.csv")
encabezado = archivo.readline().split(",")
linea = archivo.readline()
while len(linea)>0:
    datos = linea.split(",")
    x.append(float(datos[0]))
    y.append(float(datos[1]))
    linea = archivo.readline()
```

```python
m = 1.5832
b = -0.000289674

n = np.linspace(0,100)
puntos = m*n -b

plt.plot(x, y, "o")
plt.plot(n, puntos, "o", color="red")
plt.show()
```
